```python

T0 = 50.0  # temperature gradient, C
D = 6.25e-6  # thermal diffusivity, m^2/s
L = 1.0  # length of rod, m

# input parameters
dx = L*0.001        #
#dt = dx**2/D*1.00  #
dt = 0.5
#tmax = L**2/D*1.0   # 
tmax = 10000.

# coefficients of the tridiagonal matrix
alpha = gamma = -D*dt/dx**2
beta = 1.0+2.0*D*dt/dx**2

# construct initial data
N = int(L/dx)
x = np.array([k*dx for k in range(N+1)])
v = np.array([20.]*(N+1))
u = np.array([20.]*(N+1))
u[0] = T0

plt.figure(figsize=(12,8),dpi=270)
plt.plot(x, u)

# preform the evolution
t = 0.
while t < tmax:
    # swap u and v
    (u, v) = (v, u)

    # boundary conditions
    u[0] = T0
    u[N] = 0.0

    # set the j=1 and j=N-1 points of v to the correct values
    v[1] -= alpha*u[0]
    v[N-1] -= gamma*u[N]

    # forward sweep
    u[1] = v[1]/beta
    v[1] = gamma/beta
    for j in range(2, N):
        den = beta-alpha*v[j-1]
        u[j] = (v[j]-alpha*u[j-1])/den
        v[j] = gamma/den
    # backward sweep
    for j in reversed(range(1, N-1)):
        u[j] -= u[j+1]*v[j]
    t += dt
    if np.abs(t-10.) < dt: 
        plt.plot(x, u)
    if np.abs(t-100.) < dt:
        plt.plot(x,u)
    if np.abs(t-1000.) < dt:
        plt.plot(x,u)
    if np.abs(t-3000.) < dt:
        plt.plot(x,u)
    if np.abs(t-10000.) < dt:
        plt.plot(x,u)

#plt.plot(x, u)
l_list = ["$t=0$ s", "$t=10$ s", "$t=100$ s", "$t=1000$ s", "$t=3000$ s",'$t=10000$ s']
plt.legend(l_list)
plt.ylabel("$u(x, t)$ (Celcius Degree)")
plt.xlabel("$x$ (m)")
plt.show()
```

```python
import mpl_toolkits.mplot3d
eps = 1e-5  # fractional error allowed
L = 1.0  # length of each side
N = 202 #

dy = dx = L/(N-1.0)
x = np.array(range(N))*dx
y = np.array(range(N))*dy
(x, y) = np.meshgrid(x, y)
u = np.zeros((N, N))
# boundary conditions
for j in range(int(N/4),int(3*N/4)):
    u[j,int(3*N/4)] = 3.0
    u[j,int(N/4)] = -3.0
# compute over-relaxation parameter
omega = 2.0/(1.0+math.sin(math.pi*dx/L)) 

# 
white = [(j, k) for j in range(1, N-1) for k in range(1, N-1)]
# 
black = [(j, k) for j in range(int(N/4),int(3*N/4)) for k in [int(N/4),int(3*N/4)]]
# 
space = white.copy()
for i in range(len(space)):
    if white[i] in black:
        space[i] = 0
space = np.array(space)
space = space[space!=0]

n = 0  # number of iterations
err = 1.0  # average error per site
while err > eps:   # 
    # next iteration in refinement
    n = n+1
    err = 0.0
    for (j, k) in space:  # loop over white pixels then black pixels
        du = (u[j-1,k]+u[j+1,k]+u[j,k-1]+u[j,k+1])/4.0-u[j,k]
        u[j,k] += omega*du
        err += abs(du)
    err /= N**2

# surface plot of final solution, 
#fig = plt.figure(figsize=(12,8),dpi=1080)
#axis = fig.gca(projection='3d', azim=-60, elev=20)
#surf = axis.plot_surface(x, y, u.T, rstride=1, cstride=1, linewidth=0,
#                         cmap=plt.cm.jet)
#wire = axis.plot_wireframe(x, y, u.T, rstride=1+N//50, cstride=1+N//50,
#                           linewidth=0.25)
#axis.contour(x, y, u.T, 10, zdir='z', offset=-1.0)
#axis.set_xlabel('$x$(m)')
#axis.set_ylabel('$y$(m)')
#axis.set_zlabel('$u$(V)')
#axis.set_zlim(-4.0, 4.0)
```

```python
plt.figure(figsize=(12,8),dpi=270)
plt.imshow(u.T,origin='lower',extent=(0.0,L,0.0,L),cmap='jet')
plt.colorbar()
plt.xlabel('$x$(m)')
plt.ylabel('$y$(m)')
plt.title('t steps = %6f'%n)
plt.show()
```

```python
# surface plot of final solution
fig = plt.figure(figsize=(12,8),dpi=270)
axis = fig.gca(projection='3d', azim=-60, elev=20)
surf = axis.plot_surface(x, y, u.T, rstride=1, cstride=1, linewidth=0,
                         cmap=plt.cm.jet)
wire = axis.plot_wireframe(x, y, u.T, rstride=1+N//50, cstride=1+N//50,
                           linewidth=0.25)
axis.contour(x, y, u.T, 10, zdir='z', offset=-1.0)
axis.set_xlabel('$x$(m)')
axis.set_ylabel('$y$(m)')
axis.set_zlabel('$u$(V)')
axis.set_zlim(-4.0, 4.0)
fig.colorbar(surf)
plt.show()
```
